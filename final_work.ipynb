```python
df = pd.read_csv('/kc-house-data (1).csv')
df.head()
```

1.2 Изучение стоимости недвижимости

```python
plt.figure(figsize=(8, 8))
plt.hist(df['price'], bins=20)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Стоимость недвижимости')
plt.ylabel('Количество недвижимости')
plt.grid()
```

Больше всего недвижимости с ценой более 10000.

1.3 Изучите распределение квадратуры жилой

```python
plt.hist(df['sqft_living'], bins=10, color='skyblue', edgecolor='black')
plt.title('Распределение жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Частота');
```

Чаще всего встречается жилая площадь размером около 2000.
